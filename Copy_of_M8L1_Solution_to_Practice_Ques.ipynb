Q1 Write a function average_of_list(nums) that takes a list of numbers as input and returns their average. If the list is empty, return None.

```python
def average_of_list(nums):
    if not nums:
        return None
    return sum(nums) / len(nums)

print(average_of_list([10, 20, 30]))  # Output: 20.0
print(average_of_list([]))           # Output: None
```

Q2 Write a function remove_occurrences(lst, value) that removes all instances of value from the list lst

```python
def remove_occurrences(lst, value):
    return [x for x in lst if x != value]

print(remove_occurrences([1, 2, 3, 2, 4], 2))  # Output: [1, 3, 4]
```

Q3 Write a function extract_domains(emails) that takes a list of email addresses and returns a list of domain names.

```python
def extract_domains(emails):
    domains = []
    for email in emails:
        parts = email.split('@')
        if len(parts) == 2:
            domains.append(parts[1])
    return domains

emails = ['john@example.com', 'alice@work.org']
print(extract_domains(emails))  # Output: ['example.com', 'work.org']
```

Q4 Write a function slice_list(lst, start, end) that mimics list slicing but does not use the slicing operator [:].



```python
def slice_list(lst, start, end):
    result = []
    index = start
    while index < end and index < len(lst):
        result.append(lst[index])
        index += 1
    return result

print(slice_list([10, 20, 30, 40, 50], 1, 4))  # Output: [20, 30, 40]
```

Q5 Write a function word_count(sentence) that returns the number of words in a sentence. Ignore extra spaces and handle multiple spaces gracefully.

```python
def word_count(sentence):
    words = sentence.strip().split()
    return len(words)

print(word_count("  This   is a test sentence "))  # Output: 5
```
