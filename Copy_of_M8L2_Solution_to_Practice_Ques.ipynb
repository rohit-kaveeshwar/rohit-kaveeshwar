Q1 Write a function unique_words(words) that takes a list of strings and returns a set of unique words.

```python
def unique_words(words):
    return set(words)

print(unique_words(['apple', 'banana', 'apple', 'orange']))
```

Q2 Write a function common_elements(list1, list2) that returns a set containing elements present in both lists.

```python
def common_elements(list1, list2):
    return set(list1) & set(list2)

print(common_elements([1, 2, 3], [2, 3, 4]))
```

Q3 Write a function exclusive_items(a, b) that returns the elements that are present in either of the two input sets, but not in both (symmetric difference).

```python
def exclusive_items(a, b):
    return a ^ b

print(exclusive_items({1, 2, 3}, {3, 4, 5}))
```

Q4 Write a function are_disjoint(set1, set2) that returns True if the two sets have no elements in common, otherwise False.

```python
def are_disjoint(set1, set2):
    return set1.isdisjoint(set2)

print(are_disjoint({1, 2}, {3, 4}))   # True
print(are_disjoint({1, 2}, {2, 3}))   # False
```
